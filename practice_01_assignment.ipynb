## üìù Ïã§Ïäµ Î¨∏Ï†ú 2: ÏÇ∞Ìè¨ÎèÑ Ï∏°ÎèÑ

### Î∞∞Í≤Ω
**COMB (L/100 km)** ÏôÄ **ENGINE SIZE**Ïùò ÏÇ∞Ìè¨Î•º Î∂ÑÏÑùÌï©ÎãàÎã§.

### Ïù¥Î°† Î≥µÏäµ

| ÏßÄÌëú | Í≥ÑÏÇ∞ Î∞©Ïãù | Ïù¥ÏÉÅÏπò ÎØºÍ∞êÎèÑ |
|------|----------|:------------:|
| Î∂ÑÏÇ∞/ÌëúÏ§ÄÌé∏Ï∞® | ÌèâÍ∑† Í∏∞Ï§Ä Ìé∏Ï∞® Ï†úÍ≥± | ÎØºÍ∞ê |
| Î≤îÏúÑ | ÏµúÎåÄ - ÏµúÏÜå | Îß§Ïö∞ ÎØºÍ∞ê |
| IQR | Q3 - Q1 | Í∞ïÍ±¥ |

### Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Î∞©Î≤ï
- **IQR Î∞©Î≤ï**: Q1 - 1.5√óIQR ÎØ∏Îßå ÎòêÎäî Q3 + 1.5√óIQR Ï¥àÍ≥º
- **Z-score Î∞©Î≤ï**: |Z| > 2 ÎòêÎäî 3

### ÏàòÌñâÌï† ÏûëÏóÖ
1. Îëê Î≥ÄÏàòÏùò Î∂ÑÏÇ∞, ÌëúÏ§ÄÌé∏Ï∞®, Î≤îÏúÑ, IQR Í≥ÑÏÇ∞
2. Î≥ÄÎèôÍ≥ÑÏàò(CV) Í≥ÑÏÇ∞ÌïòÏó¨ ÏÉÅÎåÄÏ†Å ÏÇ∞Ìè¨ ÎπÑÍµê
3. IQR Î∞©Î≤ïÍ≥º Z-score Î∞©Î≤ïÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Î∞è ÎπÑÍµê
4. 2√ó2 ÏãúÍ∞ÅÌôî: Í∞Å Î≥ÄÏàòÏùò ÌûàÏä§ÌÜ†Í∑∏Îû® + Ï∞®ÎüâÎì±Í∏âÎ≥Ñ Î∞ïÏä§ÌîåÎ°Ø
5. Ï∞®ÎüâÎì±Í∏âÎ≥Ñ(COMPACT, MID-SIZE, SUV) Îëê Î≥ÄÏàòÏùò ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê
6. Í≤∞Í≥º Ìï¥ÏÑù

```python
# TODO 1: Îëê Î≥ÄÏàòÏùò ÏÇ∞Ìè¨ Ï∏°ÎèÑ Í≥ÑÏÇ∞
class IQR_CALCULATOR: 
    def __init__(self, data, col):
        self.data = data
        self.col = col
        
    def Q1_CAL(self):
        return self.data[self.col].quantile(0.25)
        
    def Q2_CAL(self):
        return self.data[self.col].quantile(0.5)
    
    def Q3_CAL(self):
        return self.data[self.col].quantile(0.75)
    
    def IQR_CAL(self):
        q1 = self.Q1_CAL()
        q3 = self.Q3_CAL()
        return q3 - q1 

COMB_VAR = np.var(df['COMB (L/100 km)'])
COMB_STD = np.std(df['COMB (L/100 km)'])
COMB_IQR = IQR_CALCULATOR(df,'COMB (L/100 km)')



ES_VAR = np.var(df['ENGINE SIZE'])
ES_STD = np.std(df['ENGINE SIZE'])
ES_IQR = IQR_CALCULATOR(df,'ENGINE SIZE')

print('\n', '='*30, '\n')
print(f'COMB Î∂ÑÏÇ∞ : {COMB_VAR:.2f}')
print(f'COMB ÌëúÏ§ÄÌé∏Ï∞® : {COMB_STD:.2f}')
print(f'COMB IQR: {COMB_IQR.IQR_CAL():.2f}')
print('\n', '='*30, '\n')

print('\n', '='*30, '\n')
print(f'ENGINE_SIZE Î∂ÑÏÇ∞ : {ES_VAR:.2f}')
print(f'ENGINE_SIZE ÌëúÏ§ÄÌé∏Ï∞® : {ES_STD:.2f}')
print(f'ENGINE_SIZE IQR: {ES_IQR.IQR_CAL():.2f}')
print('\n', '='*30, '\n')


# TODO 2: Î≥ÄÎèôÍ≥ÑÏàò(CV) Í≥ÑÏÇ∞
COMB_MEAN = df['COMB (L/100 km)'].mean()
COMB_CV = COMB_STD / COMB_MEAN * 100

ES_MEAN = df['ENGINE SIZE'].mean()
ES_CV = ES_STD / ES_MEAN * 100

print('\n', '='*30, '\n')
print(f'COMB CV :{COMB_CV:.2f}%')
print(f'ENGINE_SIZE CV :{ES_CV:.2f}%')
print('\n', '='*30, '\n')


# TODO 3: Ïù¥ÏÉÅÏπò ÌÉêÏßÄ (IQR, Z-score)
# Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Ìï®Ïàò (IQR Î∞©Ïãù)
def OUTLIER_IQR(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    low_outlier = Q1 - IQR * 1.5
    high_outlier = Q3 + IQR * 1.5
    
    return data[
        (data[col]<low_outlier)|
        (data[col]>high_outlier)]

COMB_OUTLIERS_IQR = OUTLIER_IQR(df, 'COMB (L/100 km)')
ES_OUTLIERS_IQR = OUTLIER_IQR(df, 'ENGINE SIZE')
print('Ïó∞ÎπÑÏùò IQR Ïù¥ÏÉÅÏπò ÌÉêÏßÄ')
display(COMB_OUTLIERS_IQR.head())
print('ÏóîÏßÑÌÅ¨Í∏∞Ïùò IQR Ïù¥ÏÉÅÏπò ÌÉêÏßÄ')
display(ES_OUTLIERS_IQR.head())

# Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Ìï®Ïàò (Z-SCORE Î∞©Ïãù)
def OUTLIER_Z_SCORE(data, col, sigma):
    std = (data[col].std())
    mean = data[col].mean()
    z_score = (data[col]-mean) / std
    outliers = data[col][np.abs(z_score)>sigma]
    return data[data[col].isin(outliers)]

COMB_OUTLIERS_Z_2 = OUTLIER_Z_SCORE(df, 'COMB (L/100 km)', 2)
ES_OUTLIERS_Z_2 = OUTLIER_Z_SCORE(df, 'ENGINE SIZE', 2)
COMB_OUTLIERS_Z_3 = OUTLIER_Z_SCORE(df, 'COMB (L/100 km)', 3)
ES_OUTLIERS_Z_3 = OUTLIER_Z_SCORE(df, 'ENGINE SIZE', 3)

print('Ïó∞ÎπÑÏùò Z-score Ïù¥ÏÉÅÏπò ÌÉêÏßÄ')
display(COMB_OUTLIERS_Z_2.head())
print('ÏóîÏßÑ ÌÅ¨Í∏∞Ïùò Z-score Ïù¥ÏÉÅÏπò ÌÉêÏßÄ')
display(ES_OUTLIERS_Z_2.head())

print('\n', '='*30, '\n')
print(f'COMB IQR Ïù¥ÏÉÅÏπò Í∞úÏàò: {len(COMB_OUTLIERS_IQR)}')
print(f'ENGINE SIZE IQR Ïù¥ÏÉÅÏπò Í∞úÏàò: {len(ES_OUTLIERS_IQR)}')
print('\n', '='*30, '\n')
print(f'COMB Z-SCORE Ïù¥ÏÉÅÏπò Í∞úÏàò (3ÏãúÍ∑∏Îßà): {len(COMB_OUTLIERS_Z_3)}')
print(f'ENGINE SIZE Z-SCORE Ïù¥ÏÉÅÏπò Í∞úÏàò(3ÏãúÍ∑∏Îßà): {len(ES_OUTLIERS_Z_3)}')
print('\n', '='*30, '\n')
print(f'COMB Z-SCORE Ïù¥ÏÉÅÏπò Í∞úÏàò (2ÏãúÍ∑∏Îßà): {len(COMB_OUTLIERS_Z_2)}')
print(f'ENGINE SIZE Z-SCORE Ïù¥ÏÉÅÏπò Í∞úÏàò(2ÏãúÍ∑∏Îßà): {len(ES_OUTLIERS_Z_2)}')
print('\n', '='*30, '\n')


# TODO 4: 2√ó2 ÏãúÍ∞ÅÌôî: Í∞Å Î≥ÄÏàòÏùò ÌûàÏä§ÌÜ†Í∑∏Îû® + Ï∞®ÎüâÎì±Í∏âÎ≥Ñ Î∞ïÏä§ÌîåÎ°Ø
def histnbox(data, col, vcol, v1, v2, v3):
    v1_data = data[data[vcol]==v1]
    v2_data = data[data[vcol]==v2]
    v3_data = data[data[vcol]==v3]
    fig, axes = plt.subplots(2,2,figsize=(12,7))
    sns.histplot(data, bins='auto', ax=axes[0,0], color='steelblue')
    axes[0,0].set_title("ÌûàÏä§ÌÜ†Í∑∏Îû®", fontweight='bold')
    sns.boxplot(v1_data, x=vcol, y=col, ax=axes[0,1], width=0.2)
    axes[0,1].set_title(f"{v1}Ïùò {col} box-plot", fontweight='bold')
    sns.boxplot(v2_data, x=vcol, y=col, ax=axes[1,0], width=0.2)
    axes[1,0].set_title(f"{v2}Ïùò {col} box-plot", fontweight='bold')
    sns.boxplot(v3_data, x=vcol, y=col, ax=axes[1,1], width=0.2)
    axes[1,1].set_title(f"{v3}Ïùò {col} box-plot", fontweight='bold')
    plt.tight_layout()
    plt.show()

histnbox(df, 'COMB (L/100 km)', 'VEHICLE CLASS', 'COMPACT', 'MID-SIZE', 'SUV')
histnbox(df, 'ENGINE SIZE', 'VEHICLE CLASS', 'COMPACT', 'MID-SIZE', 'SUV')

# TODO 5: Ï∞®ÎüâÎì±Í∏âÎ≥Ñ(COMPACT, MID-SIZE, SUV) ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê
COMB_STDS = df[df['VEHICLE CLASS'].isin(weights.index)].groupby('VEHICLE CLASS')['COMB (L/100 km)'].agg({
    ('ÌëúÏ§ÄÌé∏Ï∞®', 'std')
})
display(COMB_STDS)
# TODO 6: Í≤∞Í≥º Ìï¥ÏÑù
print('Í≤∞Í≥º Ìï¥ÏÑù:')
print(f'Ïó∞ÎπÑÏùò ÌëúÏ§ÄÌé∏Ï∞®Îäî 2.91, ÏóîÏßÑ ÌÅ¨Í∏∞Ïùò ÌëúÏ§ÄÌé∏Ï∞®Îäî 1.34Îã§. --> Ïó∞ÎπÑÎäî ÏóîÏßÑ ÌÅ¨Í∏∞Î≥¥Îã§ Ìé∏Ï∞®Í∞Ä ÌÅ¨Îã§.')
print(f'Ïó∞ÎπÑÏùò IQR Î≤îÏúÑÎäî 3.6, ÏóîÏßÑ ÌÅ¨Í∏∞Ïùò IQR Î≤îÏúÑÎäî 1.9 --> Ïó∞ÎπÑÏùò Í∞íÏù¥ Îçî ÎÑìÏùÄ Î≤îÏúÑÎ°ú ÌçºÏ†∏ ÏûàÏùåÏùÑ Ïïå Ïàò ÏûàÎã§.')
print(f'Ïó∞ÎπÑÏùò CV = 26.38%, ÏóîÏßÑ ÌÅ¨Í∏∞Ïùò CV = 39.78% --> ÏóîÏßÑ ÌÅ¨Í∏∞Ïùò Î≥ÄÎèôÏÑ±Ïù¥ Îçî ÌÅºÏùÑ Ïïå Ïàò ÏûàÎã§.')
print(f'Ïó∞ÎπÑÎäî Ïù¥ÏÉÅÏπò ÏòÅÌñ•ÏùÑ ÎßéÏù¥ Î∞õÍ≤†Íµ¨ÎÇò~~')
print(f'Ïù¥ÏÉÅÏπò ÌÉêÏÉâ Í≤∞Í≥º Ï†ïÎßêÎ°ú Ïó∞ÎπÑÎäî Ïù¥ÏÉÅÏπò ÏòÅÌñ•ÏùÑ ÎßéÏù¥ Î∞õÎäî Í≤ÉÏúºÎ°ú Î≥¥ÏûÑ')
print(f'Ïù¥ÏÉÅÏπò ÌÉêÏÉâ Î∞©Î≤ï Ï§ë Ïù¥ÏÉÅÏπòÍ∞Ä ÎßéÏù¥ ÌÉêÏßÄÎêòÎäî ÏàúÏÑú -> Z-score(2ÏãúÍ∑∏Îßà) > IQR >= Z-score(3ÏãúÍ∑∏Îßà)')
print(f'Î∞ïÏä§ÌîåÎ°ØÏùÑ ÏÇ¥Ìé¥Î¥§ÏùÑ Îïå Ï∞®Îüâ Îì±Í∏â Î≥Ñ Ïó∞ÎπÑÎäî Ïù¥ÏÉÅÏπòÎ•º ÎßéÏù¥ Ìè¨Ìï®ÌïòÍ≥† ÏûàÏùåÏùÑ Ïïå Ïàò ÏûàÎã§')
print(f'Î∞ïÏä§ÌîåÎ°ØÏùÑ ÏÇ¥Ìé¥Î¥§ÏùÑ Îïå Ï∞®Îüâ Îì±Í∏â Î≥Ñ ÏóîÏßÑ ÌÅ¨Í∏∞Îäî COMPACT, MID-SIZE Îì±Í∏âÏóêÏÑú Ïù¥ÏÉÅÏπòÎ•º ÎßéÏù¥ Ìè¨Ìï®ÌïòÍ≥† ÏûàÏùåÏùÑ Ïïå Ïàò ÏûàÎã§')
```

```python
numeric_cols = ['ENGINE SIZE', 'COMB (L/100 km)', 'COMB (mpg)', 'EMISSIONS']

# TODO 1: ÏôúÎèÑ/Ï≤®ÎèÑ Í≥ÑÏÇ∞
def cal_skew(data, list):
    tmp = data[list]
    return tmp.skew()
def cal_kurt(data, list):
    tmp = data[list]
    return tmp.kurt()

print('ÏôúÎèÑ')
display(cal_skew(df, numeric_cols))

print('Ï≤®ÎèÑ')
display(cal_kurt(df, numeric_cols))

# TODO 2: Í≤∞Í≥º DataFrame Ï†ïÎ¶¨ (ÏôúÎèÑ/Ï≤®ÎèÑ Ìï¥ÏÑù Ïª¨Îüº Ìè¨Ìï®)
def analysis(skew, kurt, tol=0.1):
    interpretations = []
    for s, k in zip(skew, kurt):
        if abs(s) < tol and abs(k) < tol:
            interpretations.append('Ï†ïÍ∑úÎ∂ÑÌè¨Ïóê Í∞ÄÍπåÏõÄ(ÎåÄÏπ≠, Íº¨Î¶¨ Î≥¥ÌÜµ)')
        elif abs(s) < tol and k > tol:
            interpretations.append('ÎåÄÏπ≠(ÌèâÍ∑†‚âàÏ§ëÏïôÍ∞í), Íº¨Î¶¨ ÎëêÍ∫ºÏõÄ ‚Üí Í∑πÎã®Í∞í ÎßéÏùå')
        elif abs(s) < tol and k < -tol:
            interpretations.append('ÎåÄÏπ≠(ÌèâÍ∑†‚âàÏ§ëÏïôÍ∞í), Íº¨Î¶¨ ÏñáÏùå ‚Üí Í∑πÎã®Í∞í Ï†ÅÏùå')
        elif s < -tol and abs(k) < tol:
            interpretations.append('ÏôºÏ™Ω Íº¨Î¶¨(ÏùåÏùò ÏôúÎèÑ): ÌèâÍ∑† < Ï§ëÏïôÍ∞í')
        elif s < -tol and k > tol:
            interpretations.append('ÏôºÏ™Ω Íº¨Î¶¨ + Íº¨Î¶¨ ÎëêÍ∫ºÏõÄ: ÏôºÏ™Ω Í∑πÎã®Í∞í ÎßéÏùå')
        elif s < -tol and k < -tol:
            interpretations.append('ÏôºÏ™Ω Íº¨Î¶¨ + Íº¨Î¶¨ ÏñáÏùå: ÏôºÏ™Ω Í∑πÎã®Í∞í Ï†ÅÏùå')
        elif s > tol and abs(k) < tol:
            interpretations.append('Ïò§Î•∏Ï™Ω Íº¨Î¶¨(ÏñëÏùò ÏôúÎèÑ): ÌèâÍ∑† > Ï§ëÏïôÍ∞í')
        elif s > tol and k > tol:
            interpretations.append('Ïò§Î•∏Ï™Ω Íº¨Î¶¨ + Íº¨Î¶¨ ÎëêÍ∫ºÏõÄ: Ïò§Î•∏Ï™Ω Í∑πÎã®Í∞í ÎßéÏùå')
        else: 
            interpretations.append('Ïò§Î•∏Ï™Ω Íº¨Î¶¨ + Íº¨Î¶¨ ÏñáÏùå: Ïò§Î•∏Ï™Ω Í∑πÎã®Í∞í Ï†ÅÏùå')
    return interpretations

sk = cal_skew(df, numeric_cols)
ku = cal_kurt(df, numeric_cols)

result = pd.DataFrame({
    "ÏôúÎèÑ": sk,
    "Ï≤®ÎèÑ": ku,
    "Ìï¥ÏÑù": analysis(sk, ku)
})

display(result)


# TODO 3: 2√ó2 ÌûàÏä§ÌÜ†Í∑∏Îû®+KDE ÏãúÍ∞ÅÌôî (Ï†úÎ™©Ïóê ÏôúÎèÑ/Ï≤®ÎèÑ ÌëúÏãú, ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÏàòÏßÅÏÑ†)

def histplot_kde(data, col1):
    fig, axes = plt.subplots(2, 2, figsize=(12,7))
    for i in data[col1].unique():
        subset = data[data[col1]==i]
        sns.histplot(subset, kde=True, bins='auto', ax=axes[0,0], color='steelblue', fill=True, alpha=0.3)
    axes.legend()
    plt.tight_layout()
    plt.show()
    



# ÏÇ¨Ïã§ Ïù¥ Î¨∏Ï†úÎ•º Ïù¥Ìï¥ÌïòÏßÄ Î™ªÌïòÍ≤†ÏäµÎãàÎã§.. Ï£ÑÏÜ°Ìï©ÎãàÎã§... gg..

# TODO 4: Ïó∞Î£åÌÉÄÏûÖ(FUEL)Î≥Ñ EMISSIONS ÏôúÎèÑ/Ï≤®ÎèÑ ÎπÑÍµê
def skewnkurt(data, col, vcol):
    fig, ax = plt.subplots(figsize=(10,5))
    for i in data[col].unique():
        subset = data[data[col]==i][vcol]
        sns.kdeplot(subset, label=f"{i} (ÏôúÎèÑ: {subset.skew():.2f}, Ï≤®ÎèÑ:{subset.kurt():.2f})", ax=ax, fill=True, alpha=0.3)
    ax.set_title(f'{col}Î≥Ñ {vcol} Î∂ÑÌè¨')
    ax.set_xlabel(f'{vcol}')
    ax.set_ylabel('Î∞ÄÎèÑ')
    ax.legend()
    
    plt.tight_layout()
    plt.show()

skewnkurt(df, 'FUEL', 'EMISSIONS')

# TODO 5: Í≤∞Í≥º Ìï¥ÏÑù
```

```python
# Í≤ÄÏ¶ù ÏΩîÎìú
import numpy as np
from scipy import stats

print("\n" + "="*60)
print("Í≤ÄÏ¶ù Í≤∞Í≥º")
print("="*60)

# ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞
test_data = [4, 7, 2, 9, 1, 5, 3, 8, 6, 5]
test_odd = [3, 1, 4, 1, 5]
test_single = [42]
test_same = [7, 7, 7, 7, 7]
test_negative = [-5, -2, 0, 3, 7]
test_outlier = [1, 2, 3, 4, 100]

all_pass = True
total = 0
passed = 0

def check(name, mine, expected, tol=0.001):
    global all_pass, total, passed
    total += 1
    if mine is None:
        print(f"  ‚ùå {name}: ÎØ∏Íµ¨ÌòÑ")
        all_pass = False
    else:
        try:
            match = abs(float(mine) - float(expected)) < tol
            if match:
                passed += 1
                print(f"  ‚úÖ {name}: {float(mine):.4f}")
            else:
                all_pass = False
                print(f"  ‚ùå {name}: {float(mine):.4f} (Ï†ïÎãµ: {float(expected):.4f})")
        except:
            all_pass = False
            print(f"  ‚ùå {name}: Ïò§Î•ò Î∞úÏÉù")

# 1. ÏÇ∞Ïà†ÌèâÍ∑†
print("\n[my_mean]")
check("Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞", my_mean(test_data), np.mean(test_data))
check("ÌôÄÏàò Í∞ú", my_mean(test_odd), np.mean(test_odd))
check("Îã®Ïùº Í∞í", my_mean(test_single), np.mean(test_single))
check("ÏùåÏàò Ìè¨Ìï®", my_mean(test_negative), np.mean(test_negative))

# 2. Ï†àÏÇ¨ÌèâÍ∑†
print("\n[my_trimmed_mean]")
check("20% Ï†àÏÇ¨", my_trimmed_mean(test_outlier, 0.2), stats.trim_mean(test_outlier, 0.2))
check("10% Ï†àÏÇ¨", my_trimmed_mean(test_data, 0.1), stats.trim_mean(test_data, 0.1))
check("0% Ï†àÏÇ¨", my_trimmed_mean(test_data, 0), stats.trim_mean(test_data, 0))

# 3. Í∞ÄÏ§ëÌèâÍ∑†
print("\n[my_weighted_mean]")
check("ÌïôÏ†ê Í≥ÑÏÇ∞", my_weighted_mean([80, 90, 70], [3, 3, 2]), np.average([80, 90, 70], weights=[3, 3, 2]))
check("ÎèôÏùº Í∞ÄÏ§ëÏπò", my_weighted_mean([100, 200], [1, 1]), np.average([100, 200], weights=[1, 1]))
check("Í∞ÄÏ§ëÏπò Îã§Î¶Ñ", my_weighted_mean([10, 20, 30], [1, 2, 3]), np.average([10, 20, 30], weights=[1, 2, 3]))

# 4. Ï§ëÏïôÍ∞í
print("\n[my_median]")
check("ÏßùÏàò Í∞ú", my_median(test_data), np.median(test_data))
check("ÌôÄÏàò Í∞ú", my_median(test_odd), np.median(test_odd))
check("Îã®Ïùº Í∞í", my_median(test_single), np.median(test_single))
check("ÏùåÏàò Ìè¨Ìï®", my_median(test_negative), np.median(test_negative))

# 5. ÏµúÎπàÍ∞í
print("\n[my_mode]")
check("Í∏∞Î≥∏ (5Í∞Ä 2Î≤à)", my_mode(test_data), stats.mode(test_data, keepdims=False).mode)
check("ÌôÄÏàò (1Ïù¥ 2Î≤à)", my_mode(test_odd), stats.mode(test_odd, keepdims=False).mode)
check("ÎèôÏùº Í∞í", my_mode(test_same), stats.mode(test_same, keepdims=False).mode)

# 6. Î∂ÑÏÇ∞
print("\n[my_variance]")
check("Î™®Î∂ÑÏÇ∞ (ddof=0)", my_variance(test_data, 0), np.var(test_data, ddof=0))
check("ÌëúÎ≥∏Î∂ÑÏÇ∞ (ddof=1)", my_variance(test_data, 1), np.var(test_data, ddof=1))
check("ÎèôÏùº Í∞í ‚Üí 0", my_variance(test_same, 0), np.var(test_same, ddof=0))

# 7. ÌëúÏ§ÄÌé∏Ï∞®
print("\n[my_std]")
check("Î™®ÌëúÏ§ÄÌé∏Ï∞®", my_std(test_data, 0), np.std(test_data, ddof=0))
check("ÌëúÎ≥∏ÌëúÏ§ÄÌé∏Ï∞®", my_std(test_data, 1), np.std(test_data, ddof=1))
check("ÎèôÏùº Í∞í ‚Üí 0", my_std(test_same, 0), np.std(test_same, ddof=0))

# 8. ÌçºÏÑºÌÉÄÏùº (Nearest Rank Î∞©Ïãù)
# Ï£ºÏùò: numpyÎäî ÏÑ†ÌòïÎ≥¥Í∞Ñ ÏÇ¨Ïö©, Nearest RankÎäî ÏßÅÏ†ë Í≥ÑÏÇ∞
print("\n[my_percentile] - Nearest Rank Î∞©Ïãù")
sorted_test = sorted(test_data)  # [1, 2, 3, 4, 5, 5, 6, 7, 8, 9]
check("0% (ÏµúÏÜüÍ∞í)", my_percentile(test_data, 0), min(test_data))
check("50% (Ï§ëÏïôÍ∞í)", my_percentile(test_data, 50), sorted_test[round((len(test_data)-1) * 0.5)])
check("100% (ÏµúÎåìÍ∞í)", my_percentile(test_data, 100), max(test_data))
# 25%, 75%Îäî Nearest Rank Î∞©ÏãùÏúºÎ°ú ÏßÅÏ†ë Í≥ÑÏÇ∞
check("25%", my_percentile(test_data, 25), sorted_test[round((len(test_data)-1) * 0.25)])
check("75%", my_percentile(test_data, 75), sorted_test[round((len(test_data)-1) * 0.75)])

# 9. IQR (Nearest Rank Í∏∞Ï§Ä)
print("\n[my_iqr]")
expected_iqr = sorted_test[round((len(test_data)-1) * 0.75)] - sorted_test[round((len(test_data)-1) * 0.25)]
check("Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞", my_iqr(test_data), expected_iqr)
check("ÎèôÏùº Í∞í ‚Üí 0", my_iqr(test_same), 0)

# ÏµúÏ¢Ö Í≤∞Í≥º
print("\n" + "="*60)
print(f"Í≤∞Í≥º: {passed}/{total} ÌÜµÍ≥º ({passed/total*100:.0f}%)")
print("="*60)
if all_pass:
    print("üéâ Î™®Îì† ÌÖåÏä§Ìä∏ ÌÜµÍ≥º!")
else:
    print("‚ö†Ô∏è ÏùºÎ∂Ä ÌÖåÏä§Ìä∏ Ïã§Ìå® - Î°úÏßÅÏùÑ ÌôïÏù∏ÌïòÏÑ∏Ïöî")
```
